```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 101)
y = np.sin(x)
```

# Example

The graph of the function:

$$ y = \sin(x) $$

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x, y)
ax.set(xlabel='x', ylabel='y', title='Test')
ax.grid()
```
